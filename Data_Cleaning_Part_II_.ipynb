```python
from google.colab import drive

import pandas as pd

spotify23_path = '/content/spotify-2023.csv'
spotify24_path = '/content/SpotifySongs2024.csv'

```

Spotify 2023

```python
df = pd.read_csv(spotify23_path, header = 1)
df.head()
df.columns = df.columns.str.strip()

df['released_year'] = pd.to_numeric(df['released_year'], errors='coerce')

df_2023 = df[df['released_year'] == 2023]

df_2023 = df_2023.rename(columns={
    "artist(s)_name": "artist_name",
    "in_spotify_charts": "peak_position",
    "streams": "stream_count",
    "bpm": "tempo",
    "danceability_%": "danceability",
    "valence_%": "valence",
    "energy_%": "energy",
    "acousticness_%": "acousticness",
    "instrumentalness_%": "instrumentalness",
    "speechiness_%": "speechiness",
    "liveness_%":"liveness"
})
df_2023.head()
```

```python
columns_to_drop = ["artist_count",
    "released_day",
    "in_spotify_playlists",
    "in_apple_playlists",
    "in_apple_charts",
    "deezer_in_playlists",
    "deezer_in_charts",
    "shazam_charts",]

df_2023 = df_2023.drop(columns = columns_to_drop, errors = "ignore")
df_2023.head(1).isna().any()
```

Spotify 2024

```python
df2 = pd.read_csv(spotify24_path, header=1)
df2.head()
df2['Release Date'] = pd.to_datetime(df2['Release Date'], errors='coerce')

# Filter rows where the year is 2024
df_2024 = df2[df2['Release Date'].dt.year == 2024]
df_2024 = df_2024.rename(columns={
    "Track": "track_name",
    "Artist": "artist_name",
    "Release Date": "release_date",
    "Explicit Track": "explicit",
    "Spotify Popularity": "track_popularity",
    "All Time Rank": "peak_position",
    "Spotify Streams": "stream_count"
})

df_2024.head()
```

```python

df_2024 = df_2024.drop(columns = ["TIDAL Popularity", "Album Name", "ISRC"], errors = "ignore")

df_2024.head(1).isna().any()
df_2024.head()
```
